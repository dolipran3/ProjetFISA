```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("../data/df_venues_processed.csv", sep=";")

data
```

```python
print("Colonnes disponibles dans le dataset:")
print(data.columns.tolist())
print(f"\nNombre total de colonnes: {len(data.columns)}")
```

```python
colonnes_a_supprimer = [
    'D1',
    'D2',
    'D3',
    'D4',
    'SNCF',
    'Date',
    'Annee_et_Semaine'
]

print("Colonnes à supprimer:")
for col in colonnes_a_supprimer:
    print(f"  - {col}")
print(f"\nNombre de colonnes à supprimer: {len(colonnes_a_supprimer)}")
```

```python
data_cleaned = data.drop(columns=colonnes_a_supprimer)

print(f"Dataset original: {data.shape}")
print(f"Dataset nettoyé: {data_cleaned.shape}")
print(f"\nColonnes conservées ({len(data_cleaned.columns)}):")
print(data_cleaned.columns.tolist())
```

```python
jour_semaine_encoded = pd.get_dummies(data_cleaned['jour_semaine'], prefix='jour')

data_with_encoding = pd.concat([data_cleaned, jour_semaine_encoded], axis=1)

data_with_encoding = data_with_encoding.drop('jour_semaine', axis=1)

print("ne-hot encoding effectué!")
print(f"\nNouveaux colonnes créées:")
for col in jour_semaine_encoded.columns:
    print(f"  - {col}")
    
print(f"\nDimensions du dataset: {data_with_encoding.shape}")
data_with_encoding.head()
```

```python
print(data_with_encoding[['jour_lundi', 'jour_mardi', 'jour_mercredi', 'jour_jeudi', 'jour_vendredi']].head(10))
```

```python
output_path_encoded = '../data/df_venues_final.csv'
data_with_encoding.to_csv(output_path_encoded, sep=';', index=False)
print(f"Dataset avec one-hot encoding sauvegardé dans: {output_path_encoded}")
print(f"Dimensions finales: {data_with_encoding.shape[0]} lignes × {data_with_encoding.shape[1]} colonnes")
```
